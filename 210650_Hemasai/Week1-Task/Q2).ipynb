```python
import numpy as np
import matplotlib.pyplot as plt
x=np.arange(0,5,0.01)
plt.plot(x,[x1 for x1 in x])
plt.plot(x,[np.sin(x1) for x1 in x])
plt.plot(x,[x1**2 for x1 in x])
plt.grid(True)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('polynomial Graph')
plt.show()
```
