```python
#Task 2
import pandas as pd
df=pd.read_csv("Regression Data.csv")
print(df)
```

```python
#1
#By Volume
print(df['Volume'])
```

```python
import matplotlib.pyplot as plt
from scipy import stats
x=df["Volume"]
y=df["CO2"]
slope,intercept,r,p,std_err=stats.linregress(x,y)
plt.scatter(x,y)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np
k1=df["Volume"].values
k1=k1.reshape(36,1)
l1=df["CO2"].values
l1=l1.reshape(36,1)
t=LinearRegression()
t=t.fit(k1,l1)
l1_pred=t.predict(k1)
r1=t.score(k1,l1)
print(r1)
```

```python
#By weight
print(df["Weight"])
```

```python
x=df["Weight"]
y=df["CO2"]
slope,intercept,r,p,std_err=stats.linregress(x,y)
plt.scatter(x,y)
plt.show()
```

```python
import numpy as np
k2=df["Weight"].values
k2=k2.reshape(36,1)
l2=df["CO2"].values
l2=l2.reshape(36,1)
t=LinearRegression()
t=t.fit(k2,l2)
l2_pred=t.predict(k2)
r2=t.score(k2,l2)
print(r2)
```

```python
#2
from sklearn.model_selection import train_test_split
p=df[['Volume','Weight']]
y=df['CO2']
p_train,p_test,y_train,y_test=train_test_split(p,y,test_size=0.3,random_state=105)
t=LinearRegression()
t.fit(p_train,y_train)
y_predict=t.predict(p_test)
plt.scatter(y_test,y_predict)
plt.xlabel('CO2')
plt.ylabel('CO2 emission')
```
